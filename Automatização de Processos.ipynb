### - Passo 1: Buscar informações das ações de forma automática:


```python
%pip install yfinance
```

```python
import yfinance as yf
```

```python
ticker = input("Digite o código da ação: ")
dados = yf.Ticker(ticker).history("6mo")
```

```python
fechamento = dados.Close
```

```python
fechamento.plot()
```

### Passo 2: Criar as análises dos dados:
- cotação máxima
- cotação mínima 
- cotação atual

```python
maxima = round(fechamento.max(), 2)
minima = round(fechamento.min(), 2)
atual = round(fechamento[-1], 2)

print(f"A cotação máximo é: R${maxima}")
print(f"A cotação mínima é: R${minima}")
print(f"A cotação atual é: R${atual}")
```

### Passo 3: Enviar um e-mail para o meu gestor com os resultados das análises:
- abrir uma nova aba (ctrl + t)
- digitar o endereço do Gmail e apertar Enter
- clicar no botão "Escrever"
- digitar o destinatário , Tab
- digitar o assunto, Tab
- digitar a mensagem do e-mail
- clicar em enviar

```python
import pyautogui
import pyperclip
```

```python
# pause de 2 segundos entre as etapas
pyautogui.PAUSE = 2

# abrir uma nova aba (ctrl + t)
pyautogui.hotkey("ctrl", "t")

# digitar o endereço do Gmail e apertar Enter
pyperclip.copy("https://mail.google.com/mail/u/1/#inbox")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("enter")

#clicar no botão Escrever
pyautogui.click(x=141, y=206)

# digitar o destinatário, Tab
pyperclip.copy("rafhael.rafinha@yahoo.com.br")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

# digitar o assunto, Tab
pyperclip.copy("Análises diárias")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

# digitar a mensagem do e-mail
mensagem = f"""
Prezado Gestor,

Seguem as análises, conforme solicitado, da ação {ticker} referentes aos últimos 6 meses:

Cotação máxima: R${maxima}
Cotação mínima: R${minima}
Cotação atual: R${atual}

Qualquer dúvida estou a disposição!

Att Rafhael Domingues
"""
pyperclip.copy(mensagem)
pyautogui.hotkey("ctrl", "v")
pyautogui.click(x=1319, y=1000)
```
